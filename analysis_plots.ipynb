 1. Speedup vs Number of Nodes

```python
import matplotlib.pyplot as plt
import numpy as np

nodes = [1, 2, 3, 4, 5, 6]
runtimes = [17, 21, 24, 28, 31, 36]  


speedup = [runtimes[0] / t for t in runtimes]


plt.figure()
plt.plot(nodes, speedup, marker='o', label='Observed Speedup')
plt.plot(nodes, nodes, linestyle='--', label='Ideal Linear Speedup') 
plt.xlabel('Number of Nodes')
plt.ylabel('Speedup')
plt.title('Speedup vs Number of Nodes')
plt.grid(True)
plt.legend()
plt.show()
```

 2. Pie Chart for Map/Shuffle/Reduce Time

```python
phases = ['Map', 'Shuffle', 'Reduce']
times = [30, 40, 45]  # in seconds

plt.figure()
plt.pie(times, labels=phases, autopct='%1.1f%%', startangle=90)
plt.title('Phase-wise Job Time Distribution')
plt.show()
```

 3. CPU and Memory Usage Plot

```python
nodes = [1, 2, 3, 4, 5, 6]
cpu_usage = [1000, 1800, 2500, 3200, 4000, 4700]  # vcore-ms
memory_usage = [1, 1, 1, 1, 1, 1]  # GB per node 

# CPU Usage
plt.figure()
plt.plot(nodes, cpu_usage, marker='o', label='CPU Usage (vcore-ms)')
plt.xlabel('Number of Nodes')
plt.ylabel('CPU Usage (vcore-ms)')
plt.title('CPU Usage vs Number of Nodes')
plt.grid(True)
plt.legend()
plt.show()

# Memory Usage
plt.figure()
plt.plot(nodes, memory_usage, marker='o', color='green', label='Memory Usage (GB)')
plt.xlabel('Number of Nodes')
plt.ylabel('Memory Usage (GB)')
plt.title('Memory Usage vs Number of Nodes')
plt.grid(True)
plt.legend()
plt.show()
```

4. Bar Chart: Runtime (Docker vs Jetstream2)

```python


frameworks = ['Hadoop', 'Spark']
docker_runtime = [115, 17]
jetstream_runtime = [90, 15]

x = np.arange(len(frameworks))  # label locations
width = 0.35  # width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, docker_runtime, width, label='Docker')
bars2 = ax.bar(x + width/2, jetstream_runtime, width, label='Jetstream2')

ax.set_ylabel('Runtime (seconds)')
ax.set_title('Runtime: Docker vs Jetstream2')
ax.set_xticks(x)
ax.set_xticklabels(frameworks)
ax.legend()
plt.grid(True)
plt.show()
```

5. Hadoop vs Spark Scaling (already have it but heres a refresh)

```python

nodes = [1, 2, 3, 4, 5, 6]
hadoop_runtime = [115, 94, 88, 88, 89, 87]
spark_runtime = [17, 21, 24, 28, 31, 36]

plt.figure()
plt.plot(nodes, hadoop_runtime, marker='o', label='Hadoop')
plt.plot(nodes, spark_runtime, marker='o', label='Spark')
plt.xlabel('Number of Nodes')
plt.ylabel('Runtime (seconds)')
plt.title('Hadoop vs Spark Runtime Scaling')
plt.grid(True)
plt.legend()
plt.show()
```

6. Saturation Marker on Runtime Plot

```python


plt.figure()
plt.plot(nodes, hadoop_runtime, marker='o', label='Hadoop')
plt.plot(nodes, spark_runtime, marker='o', label='Spark')
plt.axvline(x=3, linestyle='--', color='red', label='Saturation Point (3 Nodes)')
plt.xlabel('Number of Nodes')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime with Saturation Marker')
plt.grid(True)
plt.legend()
plt.show()
```

7. Cost vs Throughput Graph

```python

platforms = ['Docker', 'Jetstream2']
costs = [5, 15]  # Cost in dollars
throughputs = [500, 700]  # Throughput in MB/s

plt.figure()
plt.scatter(costs, throughputs, marker='o')
for i, platform in enumerate(platforms):
    plt.annotate(platform, (costs[i], throughputs[i]))
plt.xlabel('Cost ($)')
plt.ylabel('Throughput (MB/s)')
plt.title('Cost vs Throughput')
plt.grid(True)
plt.show()
```
